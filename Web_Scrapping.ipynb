```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
import os
import pandas as pd 
from io import StringIO
import numpy as np
from bs4 import BeautifulSoup
```

```python
chromedriver_path = "chromedriver.exe"
```

```python
# Inicializar el servicio de ChromeDriver
service = Service(chromedriver_path)
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com")
```

```python
URL = "https://www.exploit-db.com/"
```

```python
driver.get(URL)
```

```python
def get_html(url, selector, sleep_time=20):
    driver.get(url)
    time.sleep(sleep_time)
    element = driver.find_element(By.CSS_SELECTOR, selector) 
    return element.get_attribute("outerHTML")
```

```python
html_content = get_html(URL, "#app")
```

```python
soup = BeautifulSoup(html_content, "html.parser")
```

```python
a_tags = soup.find_all("a", href=lambda href: href and "/exploits/" in href)
```

```python
#a_tags
```

```python
titulos = []

for a in soup.find_all("a", href=True):
    if "/exploits/" in a["href"]:
        titulos.append(a.get_text(strip=True))

# convertir a DataFrame (columna)
df = pd.DataFrame(titulos, columns=["Titulo"])

print(df)
```

```python
plataformas = []

filas = soup.find_all("tr")

for fila in filas:
    celdas = fila.find_all("td")
    
    if len(celdas) >= 7:
        plataforma = celdas[6].get_text(strip=True)
        plataformas.append(plataforma)
df = pd.DataFrame(plataformas, columns=["Plataforma"])
print(df)
```

```python
tipo = soup.select_one("#exploits-table > tbody > tr:nth-child(1) > td:nth-child(6) > a").get_text(strip=True)

print(tipo)
```

```python
tipos = []

elementos = soup.select("#exploits-table > tbody > tr > td:nth-child(6)")

for el in elementos:
    tipo = el.get_text(strip=True)
    if tipo:
        tipos.append(tipo)

print(tipos)
```

```python
df = pd.DataFrame(tipos, columns=["Tipo"])
print(df)
```

```python
tipos = []

elementos = driver.find_elements(By.CSS_SELECTOR, "#exploits-table tbody tr td:nth-child(6)")

for el in elementos:
    tipo = el.get_attribute("innerText")
    
    if tipo:
        tipos.append(tipo.strip())

print(tipos)
```

```python
df = pd.DataFrame(tipos, columns=["Tipos"])
print(df)
```
